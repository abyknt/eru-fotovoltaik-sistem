```python
import osmnx as ox
import geopandas as gpd
import matplotlib
import matplotlib.pyplot as plt
```

```python
area = ox.geocode_to_gdf("Erciyes Universitesi,Kayseri,Türkiye")
```

```python
area.head()
```

## Erciyes Universitesi

```python
area.plot()
```

```python
# CRS'yi projeksiyonlu bir CRS'ye dönüştürme
area_proj = area.to_crs("EPSG:3857")

# Alan hesaplamasını projeksiyonlu geometri üzerinde yapma
total_area = area_proj.geometry.area.sum()

# Alanı yazdır
print("Erciyes Üniversitesi Alanı:", total_area, "metrekare")
```

```python
# Yer adını ve etiketleri belirleyin
place = "Erciyes Universitesi, Kayseri, Türkiye"
tags = {'highway': True}

# Overpass API'ye sorgu gönderin
response_json = ox.geometries_from_place(place, tags=tags)

# GeoDataFrame'i oluşturun
roads = gpd.GeoDataFrame.from_features(response_json)

# CRS'yi belirtin
roads.crs = "EPSG:4326"

if not roads.empty:
    # CRS'yi projeksiyonlu bir CRS'ye dönüştürme
    roads_proj = roads.to_crs("EPSG:3857")

    # Uzunluk hesaplamasını projeksiyonlu geometri üzerinde yapma
    total_length = roads_proj.geometry.length.sum()

    print("Toplam Yol Uzunluğu:", total_length, "metre")
```

## Erciyes Üniversitesi Yeşil Alanlar

Yeşil Alan hesabı yaparken natural = scrub,landuse=meadow olaral alındı.
Scrub= Çoğunlukla otları, otları ve geofitleri de içeren, çalıların (çalılar ve bodur ağaçlar) hakim olduğu bitki örtüsü ile karakterize edilen bir bitki topluluğudur.
Meadow=doğal veya yapay olarak düzenlenmiş, çimlerle kaplı, düşük boylu bitkilerin ve otların yoğun olduğu açık alanlardır


```python
# Sorgulanacak etiketleri belirle
tags = {"landuse": ["scrub", "meadow"],"leisure": "park"}

# Overpass API'ye sorgu gönder
response_json = ox.geometries_from_place(place, tags=tags)

# GeoDataFrame'i oluştur
green_areas = gpd.GeoDataFrame.from_features(response_json)
green_areas.head()
```

```python
green_areas.plot()
```

```python
green_areas.crs = "EPSG:4326"
if not green_areas.empty:
    # CRS'yi projeksiyonlu bir CRS'ye dönüştürme
    green_areas_proj = green_areas.to_crs("EPSG:3857")

    # Alan hesaplamasını projeksiyonlu geometri üzerinde yapma
    total_area = green_areas_proj.geometry.area.sum()

    print("Yeşil Alan:", total_area, "metrekare")
```

```python
tags = {"natural": "water"}

# Overpass API'ye sorgu gönder
response_json = ox.geometries_from_place(place, tags=tags)

# GeoDataFrame'i oluştur
water_areas = gpd.GeoDataFrame.from_features(response_json)
water_areas.head()
```

```python
water_areas.plot()
```

```python
water_areas.crs = "EPSG:4326"
if not water_areas.empty:
    # CRS'yi projeksiyonlu bir CRS'ye dönüştürme
    water_areas_proj = water_areas.to_crs("EPSG:3857")

    # Alan hesaplamasını projeksiyonlu geometri üzerinde yapma
    total_area = water_areas_proj.geometry.area.sum()

    print("Yeşil Alan:", total_area, "metrekare")
```

# Birleştirilen harita

```python
# Alan
area = ox.geocode_to_gdf(place_name)

# Kullanılan bölgeleri çiz
fig, ax = plt.subplots(figsize=(10, 10))
area.plot(ax=ax, color='lightgray', edgecolor='black')
# Binaları al
buildings = ox.geometries_from_place(place_name, tags={'building': True})
buildings.plot(ax=ax, color='blue', alpha=0.7)
roads = ox.geometries_from_place(place_name, tags={'highway': True})
roads.plot(ax=ax, color='black', alpha=0.5, linewidth=1)
parking = ox.geometries_from_place(place_name, tags={'amenity': 'parking'})
parking.plot(ax=ax, color='orange', alpha=0.5)
green_areas = ox.geometries_from_place(place_name, tags={"landuse": ["scrub", "meadow"],"leisure": "park"})
green_areas.plot(ax=ax, color='green', alpha=0.5)
water_areas = ox.geometries_from_place(place_name, tags={'natural': 'water'})
water_areas.plot(ax=ax, color='cyan', alpha=0.7)

plt.axis('off')
plt.show()
```

```python
# Alan
area = ox.geocode_to_gdf(place_name)
buildings = ox.geometries_from_place(place_name, tags={'building': True})
buildings.plot(ax=ax, color='blue', alpha=0.7)
roads = ox.geometries_from_place(place_name, tags={'highway': True})
roads.plot(ax=ax, color='black', alpha=0.5, linewidth=1)
parking = ox.geometries_from_place(place_name, tags={'amenity': 'parking'})
parking.plot(ax=ax, color='orange', alpha=0.5)
green_areas = ox.geometries_from_place(place_name, tags={"landuse": ["scrub", "meadow"],"leisure": "park"})
green_areas.plot(ax=ax, color='green', alpha=0.5)
water_areas = ox.geometries_from_place(place_name, tags={'natural': 'water'})
water_areas.plot(ax=ax, color='cyan', alpha=0.7)
# Kullanılamayacak alanları çıkart
unused_area = area.difference(roads.unary_union).difference(green_areas.unary_union).difference(water_areas.unary_union)

# Kullanılmayan alanı çiz
fig, ax = plt.subplots(figsize=(10, 10))
unused_area.plot(ax=ax, color='lightgray', edgecolor='black')
buildings.plot(ax=ax, color='blue', alpha=0.7)
parking.plot(ax=ax, color='orange', alpha=0.5)

plt.axis('off')
plt.show()
```

```python
# CRS'yi projeksiyonlu bir CRS'ye dönüştürme
unused_area_proj = unused_area.to_crs("EPSG:3857")

# Alan hesaplamasını projeksiyonlu geometri üzerinde yapma
unused_area_area = unused_area_proj.geometry.area.sum()

print("Kullanılabicelek Alan:", unused_area_area, "metrekare")
```
